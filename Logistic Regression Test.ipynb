```python
import pandas as pd 

from sklearn.model_selection import train_test_split


from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


import os
```

```python
os.chdir('C:/Downloads')

test_df = pd.read_csv('processed_test_df.csv')

test_df.dropna(how='any', inplace = True)



flag_features = test_df[[col for col in test_df.columns if col.startswith('flag')]]
non_flag_features = test_df[[col for col in test_df.columns if not col.startswith('flag')]]

# Select all columns that do NOT start with 'flag'

#First create the mask 
valid_mask = (flag_features['flag_agn'] != -99)

non_flag_features_cleaned = non_flag_features[valid_mask].reset_index(drop=True)
flag_features_cleaned = flag_features[valid_mask].reset_index(drop=True)




```

```python
non_numeric_cols = non_flag_features.select_dtypes(include=['object', 'category', 'bool']).columns
test_df[non_numeric_cols].isna().sum()
```

```python
from sklearn.multioutput import MultiOutputClassifier
from sklearn.preprocessing import StandardScaler


# Identify all non-numeric columns in non_flag_features
non_numeric_cols = non_flag_features_cleaned.select_dtypes(include=['object', 'category', 'bool']).columns

# One-hot encode all non-numeric columns (convert Index to list)
non_flag_features_encoded = pd.get_dummies(non_flag_features_cleaned, columns=non_numeric_cols.tolist())

X_train, X_test, y_train, y_test = train_test_split(non_flag_features_encoded, flag_features_cleaned)

# Remove target columns with only one unique value (required for LogisticRegression)
unique_class_cols = [col for col in y_train.columns if y_train[col].nunique() > 1]
y_train = y_train[unique_class_cols]
y_test = y_test[unique_class_cols]

#Scale Data 
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test) 

multi_model = MultiOutputClassifier(LogisticRegression(random_state=42, max_iter=200))
multi_model.fit(X_train_scaled, y_train)

y_pred = multi_model.predict(X_test_scaled)
```

```python
#Check accuracy for each label 

from sklearn.metrics import accuracy_score

#Extract the test data columns 
for i, col in enumerate(y_test.columns):
    acc = accuracy_score(y_test.iloc[:, i], y_pred[:,i])
    print(f"Accuracy for {col}: {acc:.2f}")
```
